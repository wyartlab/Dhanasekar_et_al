```python
import pyabf
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Load the ABF file
file_path = r'C:\Users\m.dhanasekar\Downloads\test.abf'  # Replace with the path to your .abf file
abf = pyabf.ABF(file_path)

# Extract the time and data (voltage trace)
time = abf.sweepX  # Time in seconds
voltage = abf.sweepY  # Voltage in mV

# Find the peaks (action potentials) in the voltage trace
peaks, _ = find_peaks(voltage, height=-20)  # Detect peaks above 0 mV

# Find the troughs (local minima) in the voltage trace
troughs, _ = find_peaks(-voltage)  # Detect troughs as negative peaks

# Calculate spike frequency (number of spikes per second)
spike_count = len(peaks)
recording_duration = time[-1] - time[0]
spike_frequency = spike_count / recording_duration if recording_duration > 0 else 0

# Calculate amplitude of each spike (peak - resting membrane potential)
resting_potential = np.mean(voltage[:int(len(voltage) * 0.1)])  # Assuming first 10% is resting
amplitudes = voltage[peaks] - resting_potential

# Calculate interspike interval (ISI)
interspike_intervals = np.diff(time[peaks])

# Calculate latency to first spike
latency_to_first_spike = time[peaks[0]] if spike_count > 0 else np.nan

# Plot the voltage trace with peaks and troughs highlighted
plt.figure(figsize=(12, 6))
plt.plot(time, voltage, label='Voltage Trace')
plt.plot(time[peaks], voltage[peaks], "ro", label='Peaks')
plt.plot(time[troughs], voltage[troughs], "ko", label='Troughs')
plt.title('Voltage Trace with Peaks and Troughs')
plt.xlim(6.2,6.7)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
plt.legend()
plt.show()

# Display results
print(f"Spike Frequency: {spike_frequency} Hz")
print(f"Amplitudes: {amplitudes}")
print(f"Interspike Intervals: {interspike_intervals}")
print(f"Latency to First Spike: {latency_to_first_spike} seconds")

# Optional: Save results to a file
with open("spike_analysis_results.txt", "w") as f:
    f.write(f"Spike Frequency: {spike_frequency} Hz\n")
    f.write(f"Amplitudes: {amplitudes}\n")
    f.write(f"Interspike Intervals: {interspike_intervals}\n")
    f.write(f"Latency to First Spike: {latency_to_first_spike} seconds\n")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import sem

# === paths ===
input_path  = r"C:\Maha\DATA\patch_clamp\charlotte\MAHA_data\loose_patch_quantification.xlsx"
output_path = r"C:\Maha\DATA\patch_clamp\charlotte\MAHA_data\latency_summary_by_file.xlsx"

# === load ===
if input_path.lower().endswith(".csv"):
    df = pd.read_csv(input_path)
else:
    df = pd.read_excel(input_path)

# Normalize column names
df.columns = [str(c).strip() for c in df.columns]

# Identify latency columns (everything after these known columns)
core_cols = ['file','mean','median','sem','num_pulses']
latency_cols = [c for c in df.columns if c not in core_cols]

if not latency_cols:
    raise ValueError("No latency columns detected. Make sure columns after 'num_pulses' are the latency values.")

# Coerce to numeric
for c in latency_cols + ['mean','median','sem','num_pulses']:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# --- compute stats from the latency values on each row (per file) ---
def row_stats(row):
    vals = pd.to_numeric(row[latency_cols], errors='coerce').dropna()
    if vals.empty:
        return pd.Series({
            "calc_mean": np.nan, "calc_median": np.nan, "calc_std": np.nan, "calc_sem": np.nan,
            "calc_q1": np.nan, "calc_q3": np.nan, "calc_IQR": np.nan, "n": 0
        })
    q1 = vals.quantile(0.25, interpolation="midpoint")
    q3 = vals.quantile(0.75, interpolation="midpoint")
    return pd.Series({
            "calc_mean": vals.mean(),
            "calc_median": vals.median(),
            "calc_std": vals.std(),
            "calc_sem": sem(vals),
            "calc_q1": q1,
            "calc_q3": q3,
            "calc_IQR": q3 - q1,
            "n": len(vals)
    })

stats_df = df.apply(row_stats, axis=1)
out = pd.concat([df[['file','num_pulses']].reset_index(drop=True), stats_df], axis=1)

# Tidy column names for the summary sheet (per-file)
summary_stats = out.rename(columns={
    'calc_mean':'mean', 'calc_median':'median', 'calc_std':'std', 'calc_sem':'sem',
    'calc_q1':'q1', 'calc_q3':'q3', 'calc_IQR':'IQR'
})

# --- Global pooled stats (ignore file/stim duration) ---
all_vals = pd.to_numeric(df[latency_cols].values.ravel(), errors='coerce')
all_vals = pd.Series(all_vals).dropna()

def series_stats(s):
    if s.empty:
        return pd.Series({"file":"ALL_files","n":0,"mean":np.nan,"median":np.nan,"std":np.nan,
                          "sem":np.nan,"q1":np.nan,"q3":np.nan,"IQR":np.nan})
    q1 = s.quantile(0.25, interpolation="midpoint")
    q3 = s.quantile(0.75, interpolation="midpoint")
    return pd.Series({
        "file": "ALL_files",
        "n": int(s.size),
        "mean": s.mean(),
        "median": s.median(),
        "std": s.std(),
        "sem": sem(s),
        "q1": q1,
        "q3": q3,
        "IQR": q3 - q1
    })

global_stats = series_stats(all_vals).to_frame().T  # one-row DataFrame

# --- Readable lines (legend-ready), including pooled line ---
def short_line(row, decimals=4):
    if int(row['n']) == 0:
        return f"{row['file']}: n=0"
    return (f"{row['file']}: n={int(row['n'])}, "
            f"mean={row['mean']:.{decimals}f}, "
            f"median={row['median']:.{decimals}f}, "
            f"std={row['std']:.{decimals}f}, "
            f"SEM={row['sem']:.{decimals}f}, "
            f"IQR={row['q1']:.{decimals}f}{row['q3']:.{decimals}f}")

readable_df = pd.DataFrame({
    "file": summary_stats['file'],
    "Summary": summary_stats.apply(short_line, axis=1)
})
# append pooled line
readable_df = pd.concat(
    [readable_df,
     pd.DataFrame({"file":["ALL_files"], "Summary":[short_line(global_stats.iloc[0]) ]})],
    ignore_index=True
)

# Optional: include your originally provided mean/median/sem for reference
provided_cols = [c for c in ['mean','median','sem'] if c in df.columns]
reference_sheet = None
if provided_cols:
    reference_sheet = pd.concat(
        [df[['file','num_pulses'] + provided_cols],
         summary_stats[['mean','median','sem']].add_prefix('calc_')],
        axis=1
    )

# Save to Excel
with pd.ExcelWriter(output_path) as writer:
    # Raw per-row table
    df.to_excel(writer, sheet_name="Raw_Table", index=False)
    # Per-file summary
    summary_stats[['file','num_pulses','n','mean','median','std','sem','q1','q3','IQR']].to_excel(
        writer, sheet_name="Summary_Stats", index=False
    )
    # Global pooled summary
    global_stats[['file','n','mean','median','std','sem','q1','q3','IQR']].to_excel(
        writer, sheet_name="Global_Stats", index=False
    )
    # Readable lines (incl. ALL_files)
    readable_df.to_excel(writer, sheet_name="Readable", index=False)
    # Provided vs Calculated
    if reference_sheet is not None:
        reference_sheet.to_excel(writer, sheet_name="Provided_vs_Calculated", index=False)

print(f"Exported: {output_path}")

```

```python
import pandas as pd
import numpy as np
from scipy.stats import sem

input_path  = r"C:\Maha\DATA\patch_clamp\charlotte\MAHA_data\num_spikes.xlsx"  # replace with your Excel/CSV
output_path = r"C:\Maha\DATA\patch_clamp\charlotte\MAHA_data\num_spikes_summary.xlsx"
# === load ===
if input_path.lower().endswith(".csv"):
    df = pd.read_csv(input_path)
else:
    df = pd.read_excel(input_path)

df.columns = [str(c).strip() for c in df.columns]

# Identify columns
core_cols = ['file','mean','median','sem','num_pulses','total','Hz','num_spikes']
spike_cols = [c for c in df.columns if c not in core_cols]

for c in spike_cols + ['mean','median','sem','num_pulses','total','Hz','num_spikes']:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# --- stats function ---
def row_stats(series):
    s = pd.to_numeric(series, errors='coerce').dropna()
    if s.empty:
        return pd.Series({
            "mean": np.nan, "median": np.nan, "std": np.nan, "sem": np.nan,
            "q1": np.nan, "q3": np.nan, "IQR": np.nan, "n": 0
        })
    q1 = s.quantile(0.25, interpolation="midpoint")
    q3 = s.quantile(0.75, interpolation="midpoint")
    return pd.Series({
        "mean": s.mean(),
        "median": s.median(),
        "std": s.std(),
        "sem": sem(s),
        "q1": q1,
        "q3": q3,
        "IQR": q3 - q1,
        "n": len(s)
    })

# --- Per-stim summaries ---
rowwise = []
for _, row in df.iterrows():
    stats = row_stats(row[spike_cols])
    stats["file"] = row["file"]
    stats["num_pulses"] = row["num_pulses"]
    stats["total"] = row["total"]
    stats["Hz"] = row["Hz"]
    stats["num_spikes"] = row["num_spikes"]
    rowwise.append(stats)

summary_stats = pd.DataFrame(rowwise)

# --- Global pooled summary (ignore stim) ---
all_vals = pd.to_numeric(df[spike_cols].values.ravel(), errors='coerce')
all_vals = pd.Series(all_vals).dropna()

global_stats = row_stats(all_vals)
global_stats["file"] = "ALL_stim_durations"
global_stats = pd.DataFrame([global_stats])

# --- Readable lines ---
def short_line(row, decimals=3):
    if int(row['n']) == 0:
        return f"{row['file']}: n=0"
    return (f"{row['file']}: n={int(row['n'])}, "
            f"mean={row['mean']:.{decimals}f}, "
            f"median={row['median']:.{decimals}f}, "
            f"std={row['std']:.{decimals}f}, "
            f"SEM={row['sem']:.{decimals}f}, "
            f"IQR={row['q1']:.{decimals}f}{row['q3']:.{decimals}f}")

readable_df = pd.DataFrame({
    "file": summary_stats['file'],
    "Summary": summary_stats.apply(short_line, axis=1)
})
# add global line
readable_df = pd.concat([readable_df,
                         pd.DataFrame({"file":["ALL_stim_durations"],
                                       "Summary":[short_line(global_stats.iloc[0])]})],
                        ignore_index=True)

# --- Save ---
with pd.ExcelWriter(output_path) as writer:
    df.to_excel(writer, sheet_name="Raw_Table", index=False)
    summary_stats[['file','num_pulses','total','Hz','num_spikes','n','mean','median','std','sem','q1','q3','IQR']].to_excel(
        writer, sheet_name="Summary_Stats", index=False)
    global_stats[['file','n','mean','median','std','sem','q1','q3','IQR']].to_excel(
        writer, sheet_name="Global_Stats", index=False)
    readable_df.to_excel(writer, sheet_name="Readable", index=False)

print(f"Exported: {output_path}")
```
