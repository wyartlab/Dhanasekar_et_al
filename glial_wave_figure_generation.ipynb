```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import sem, pearsonr

# --- Load data ---
file_path = r'C:\Maha\DATA\dbh_activation_gfap_imaging_vnr\071223\frames_replaced\extraction\concatenated_new.csv'
data = pd.read_csv(file_path)

# --- Clean column names ---
data.rename(columns={
    'Value_Time_to_peak(s)': 'Value_Time_to_peak_s',
    'Value_Amplitude (df/f)': 'Value_Amplitude_df_f',
    'Value_Slope_max ((df/f)/s)': 'Value_Slope_max_df_f_s',
    'Value_Slope_half (df/f)/s)': 'Value_Slope_half_df_f_s',
    'Value_Half_width (s)': 'Value_Half_width_s',
    'Value_Tau (s)': 'Value_Tau_s'
}, inplace=True)

# --- Define ROI order ---
data['ROI'] = data['Rank_time_to_peak'].astype(str)
roi_order = ['1', '2', '3', '4', '5', '6']
data['ROI'] = pd.Categorical(data['ROI'], categories=roi_order, ordered=True)
data['ROI_num'] = data['ROI'].astype(int)

# --- Final cluster color palette (C1C6) ---
roi_colors = {
    '1': '#F38CAB',  # rose-pink
    '2': '#E754A5',  # pink-magenta
    '3': '#C57AE0',  # lilac-violet
    '4': '#B3A8F9',  # periwinkle-lavender
    '5': '#6A93F2',  # sky blue
    '6': '#4359D0',  # deep cobalt-blue
}
palette_list = [roi_colors[r] for r in roi_order]

# --- Compute group means for regression ---
grp = data.groupby('ROI')['Value_Time_to_peak_s']
means = grp.mean().reindex(roi_order)
errors = grp.apply(sem).reindex(roi_order)

# --- Plot ---
sns.set(style='white', context='talk')
fig, ax = plt.subplots(figsize=(6, 6))

# Narrow colored boxplots
sns.boxplot(
    data=data, x='ROI', y='Value_Time_to_peak_s',
    order=roi_order,
    width=0.4,
    palette=roi_colors,
    showcaps=True,
    boxprops={'edgecolor': 'black', 'linewidth': 1},
    medianprops={'color': 'black', 'linewidth': 2},
    whiskerprops={'color': 'black'},
    capprops={'color': 'black'},
    ax=ax
)

# Overlay stripplot (individual points)
sns.stripplot(
    data=data, x='ROI', y='Value_Time_to_peak_s',
    order=roi_order,
    palette=roi_colors,
    size=5, jitter=0.2, alpha=0.75, ax=ax
)

# --- Linear regression (based on cluster means) ---
r, p = pearsonr(means.index.astype(int), means.values)
coef = np.polyfit(means.index.astype(int), means.values, 1)
poly = np.poly1d(coef)
x_line = np.linspace(1, 6, 200)
y_line = poly(x_line)
ax.plot(
    x_line - 1, y_line,
    linestyle='--', color='black', linewidth=2,
    label=f'Linear fit (r={r:.2f}, p={p:.3g})'
)

# --- Cosmetics ---
ax.set_xticks(range(6))
ax.set_xticklabels(roi_order)
ax.set_xlabel('Cluster (C1C6)')
ax.set_ylabel('Time to Peak (s)')
ax.set_title('Time to Peak Across Clusters (C1C6)', fontsize=16)
ax.legend(frameon=False, loc='upper left')
ax.set_ylim(bottom=0)

# Clean up frame
for spine in ax.spines.values():
    spine.set_visible(False)
ax.grid(False)

plt.tight_layout()

# --- Save if needed ---
out_svg = r'C:\Maha\DATA\Dhanasekar_et_al_2025\figures\drafts\glial_calcium_imaging\SVGs\time_to_peak_colored_boxplots.svg'
# out_pdf = r'C:\Maha\DATA\...\time_to_peak_colored_boxplots.pdf'
fig.savefig(out_svg)
# fig.savefig(out_pdf)

plt.show()

```
